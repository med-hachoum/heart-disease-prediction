## GetMarkets take one var , exemple :GetMarkets('BTC')

```python
def GetMarkets(coin):
    api_curl = 'https://www.cryptopia.co.nz/api/GetMarkets/{}'.format(coin)
    api_call = urllib2.urlopen(api_curl)
    api_response =json.load(api_call)
    if len(api_response['Data'])==0:
        return False
    else:
        return api_response
```

<h3><span style="color: #339966;"><strong><span style="color: #333300;">Markets_infos: take two vars</span>(<span style="color: #ff6600;">markets</span>,<span style="color: #ffff00;">info</span>)</strong></span></h3>
<h3><span style="color: #339966;"><strong><span style="color: #ff6600;">markets</span> : <span style="color: #333300;">results of GetMarkets function &nbsp;</span></strong></span></h3>
<h3><span style="color: #ffff00;"><strong>info :</strong></span><span style="color: #339966;"><strong>&nbsp;<span style="color: #333300;">variables :</span></strong></span></h3>
<p><span style="color: #003300;">'BaseVolme' , 'BidPrice' , 'ByBaseVolme' , 'ByVolme' , 'Change' , 'Close' , 'High' , 'Label' , 'LastPrice' , 'Low' , 'Open' , 'SellBaseVolme' , 'SellVolme' , 'TradePairId' , 'Volme'</span></p>

```python
def Markets_infos(markets,info):
    info_list=[]
    for i in range(len(markets['Data'])):
        info_list.append(markets['Data'][i][info]) 
    return info_list
    
```

```python
xxx =GetMarkets('BTC')
```

```python
for l,p in zip(Markets_infos(xxx,'Label'),Markets_infos(xxx, 'LastPrice')) :
    print l,p
```
